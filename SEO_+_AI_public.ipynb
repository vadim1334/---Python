```python
cms_signatures = {
    'WordPress': {
        'patterns': [
            re.compile(r'wp-content/themes', re.I),
            re.compile(r'wp-includes', re.I),
            re.compile(r'wp-json', re.I),
            re.compile(r'wp-admin', re.I),
            re.compile(r'wordpress', re.I)
        ]
    },
    'Bitrix': {
        'patterns': [
            re.compile(r'bitrix', re.I),
            re.compile(r'bx\.js', re.I),
            re.compile(r'/bitrix/templates', re.I),
            re.compile(r'BX\.(message|ajax)', re.I)
        ]
    },
    'Tilda': {
        'patterns': [
            re.compile(r'tilda\.js', re.I),
            re.compile(r'tilda\.css', re.I),
            re.compile(r'data-tilda', re.I),
            re.compile(r'tildacdn', re.I)
        ]
    }
}

def fallback_detection(html):
    for cms, config in cms_signatures.items():
        matches = [p for p in config.get('patterns', []) if p.search(html)]
        if matches:
            confidence = min((len(matches) / len(config['patterns'])) * 100, 100)
            if confidence > 30:
                return {'name': cms, 'confidence': confidence, 'detected_by': 'patterns'}
    return None
```

```python
#—Ü–∏–∫–ª –ø—Ä–æ–≤–µ—Ä–∫–∏
def check_ssl(domain):
    context = ssl.create_default_context()
    conn = context.wrap_socket(
        socket.socket(socket.AF_INET),
        server_hostname=domain,
    )
    try:
        conn.settimeout(3.0)
        conn.connect((domain, 443))
        cert = conn.getpeercert()
        expires = datetime.datetime.strptime(cert['notAfter'], '%b %d %H:%M:%S %Y %Z')
        days_left = (expires - datetime.datetime.utcnow()).days
        return {
            'valid': True,
            'issuer': cert['issuer'][0][0][1],
            'expiresAt': expires.strftime('%Y-%m-%d'),
            'daysUntilExpiration': days_left
        }
    except Exception as e:
        return {'valid': False, 'error': str(e)}

def check_dns(domain):
    try:
        ip = socket.gethostbyname(domain)
        return {'A': ip}
    except Exception as e:
        return {'A': None, 'error': str(e)}

def detect_cdn(domain):
    try:
        ip = socket.gethostbyname(domain)
        cdn_hints = ['Cloudflare', 'Akamai', 'Fastly', 'Amazon', 'Google']
        #–ø—Ä–∏–º–µ—Ä –ø—Ä–æ—Å—Ç–æ–π —ç–≤—Ä–∏—Å—Ç–∏–∫–∏
        if ip.startswith('104.') or ip.startswith('172.') or ip.startswith('13.'):
            return 'Possibly behind CDN'
        return 'None detected'
    except:
        return 'Unknown'

def measure_performance(url):
    try:
        start = time.time()
        res = requests.get(url)
        end = time.time()
        return {
            'loadTime': round((end - start) * 1000, 2),  # ms
            'status_code': res.status_code,
            'httpVersion': res.raw.version
        }
    except Exception as e:
        return {'error': str(e)}

def check_compression(url):
    try:
        res = requests.get(url, headers={'Accept-Encoding': 'gzip, br'})
        encoding = res.headers.get('Content-Encoding')
        return {
            'enabled': encoding is not None,
            'type': encoding or 'none'
        }
    except Exception as e:
        return {'enabled': False, 'error': str(e)}

def tech_analysis(url):
    domain = urlparse(url).hostname
    print(f'üîç –ê–Ω–∞–ª–∏–∑–∏—Ä—É–µ–º: {domain}...\n')

    return {
        'ssl': check_ssl(domain),
        'dns': check_dns(domain),
        'cdn': detect_cdn(domain),
        'performance': measure_performance(url),
        'compression': check_compression(url),
    }
```

```python
#—Ü–∏–∫–ª
from urllib.parse import urljoin
import requests

def check_redirects(url, max_redirects=10):
    redirects = []
    current_url = url
    redirect_count = 0

    while redirect_count < max_redirects:
        try:
            response = requests.head(current_url, allow_redirects=False, timeout=5)
        except requests.exceptions.RequestException as e:
            redirects.append({
                'url': current_url,
                'status': 'ERROR',
                'error': str(e)
            })
            break

        redirects.append({
            'url': current_url,
            'status': response.status_code,
            'headers': dict(response.headers)
        })

        if 300 <= response.status_code < 400:
            location = response.headers.get('Location')
            if not location:
                break
            current_url = urljoin(current_url, location)
            redirect_count += 1
        else:
            break

    return redirects


def print_redirects(redirects):
    for i, step in enumerate(redirects, start=1):
        status = step.get("status")
        url = step.get("url")
        if status == "ERROR":
            print(f"{i}. {url} ‚Üí ‚ùå –û—à–∏–±–∫–∞: {step.get('error')}")
        else:
            print(f"{i}. {url} ‚Üí {status}")
            if 300 <= status < 400:
                location = step["headers"].get("Location")
                if location:
                    print(f"   ‚îî‚îÄ‚îÄ –ø–µ—Ä–µ–Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–µ –Ω–∞ {location}")
```

```python
def fetch_html_limited(url, max_size=2 * 1024 * 1024, timeout=30):
    """
    –ó–∞–≥—Ä—É–∂–∞–µ—Ç HTML —Å —Å–∞–π—Ç–∞ —Å –ª–∏–º–∏—Ç–æ–º –ø–æ –æ–±—ä—ë–º—É (max_size –±–∞–π—Ç).
    –ü—Ä–µ—Ä—ã–≤–∞–µ—Ç –∑–∞–≥—Ä—É–∑–∫—É, –µ—Å–ª–∏ –ø—Ä–µ–≤—ã—à–µ–Ω –ª–∏–º–∏—Ç.
    """
    try:
        response = requests.get(url, stream=True, timeout=timeout)
        response.raise_for_status()

        chunks = []
        total_length = 0

        for chunk in response.iter_content(chunk_size=8192, decode_unicode=True):
            if chunk:
                chunks.append(chunk)
                total_length += len(chunk)
                if total_length > max_size:
                    #–ø—Ä–µ—Ä—ã–≤–∞–µ–º –∑–∞–≥—Ä—É–∑–∫—É ‚Äî –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –¥–∞–Ω–Ω—ã—Ö –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞
                    break

        html = ''.join(chunks)
        return html

    except requests.RequestException as e:
        print(f"–û—à–∏–±–∫–∞ –ø—Ä–∏ –∑–∞–≥—Ä—É–∑–∫–µ: {e}")
        return None
```

```python
SEED_KEYWORDS = [
    "",
    "",
    "",
    "",
    ""
]

TOP_N_KEYWORDS = 30
TOP_N_FOR_PROMPT = 20


WORDSTAT_TOKEN = ''
CLIENT_LOGIN   = ''    #–æ–±—è–∑–∞—Ç–µ–ª—å–Ω–æ –¥–ª—è Wordstat API

import requests, time, pandas as pd, unicodedata, json, re, itertools

API_URL = ''

def _norm(s):
    s = unicodedata.normalize("NFKC", s).strip().lower()
    return re.sub(r"\s+", " ", s)

def ws_request(payload):
    headers = {
        "Authorization": f"Bearer {WORDSTAT_TOKEN}",
        "Client-Login": CLIENT_LOGIN,
        "Accept-Language": "ru",
        "Content-Type": "application/json; charset=utf-8",
    }
    r = requests.post(API_URL, json=payload, headers=headers, timeout=45)
    data = r.json()
    if "error" in data:
        raise RuntimeError(f"Wordstat API error: {data['error']['error_code']} {data['error']['error_detail']}")
    return data

def create_report(phrases):
    payload = {"method": "create", "params": {"Phrases": phrases}}
    data = ws_request(payload)
    return data["data"]["Id"]

def get_report(id_):
    payload = {"method": "get", "params": {"ReportID": id_}}
    data = ws_request(payload)
    return data["data"]

def delete_report(id_):
    payload = {"method": "delete", "params": {"ReportID": id_}}
    try:
        ws_request(payload)
    except:
        pass

def wait_ready(id_, timeout=180):
    t0 = time.time()
    while time.time() - t0 < timeout:
        try:
            rep = get_report(id_)
            return rep
        except:
            time.sleep(2)
    raise TimeoutError("–û—Ç—á—ë—Ç Wordstat –Ω–µ —Å—Ç–∞–ª –≥–æ—Ç–æ–≤—ã–º.")

def aggregate(rows):
    freq = {}
    for row in rows:
        for block in ("SearchedWith", "SearchedAlso"):
            for item in row.get(block, []):
                kw = _norm(item["Phrase"])
                shows = int(item.get("Shows", 0))
                if kw and not re.fullmatch(r"[0-9\-\s]+", kw):
                    freq[kw] = freq.get(kw, 0) + shows
    df = pd.DataFrame([{"keyword": k, "shows": v} for k, v in freq.items()])
    return df.sort_values("shows", ascending=False).reset_index(drop=True) if not df.empty else df

def fallback(seeds):
    mods = ["–ø—Ä–∏–º–µ—Ä", "–∏–Ω—Ç–µ–≥—Ä–∞—Ü–∏—è", "–≤–Ω–µ–¥—Ä–µ–Ω–∏–µ", "—Ä–µ—à–µ–Ω–∏–µ", "enterprise", "–æ–Ω-–ø—Ä–µ–º–∏—Å", "–¥–ª—è –±–∏–∑–Ω–µ—Å–∞"]
    out = set(_norm(s) for s in seeds)
    for s in seeds:
        b = _norm(s)
        for m in mods:
            out.add(f"{b} {m}")
            out.add(f"{m} {b}")
    return list(out)[:TOP_N_KEYWORDS]

# ---- RUN ----
try:
    seeds = [_norm(x) for x in SEED_KEYWORDS if _norm(x)]
    report_id = create_report(seeds)
    data = wait_ready(report_id)
    delete_report(report_id)

    df = aggregate(data)
    if df.empty:
        print("‚ö†Ô∏è API –æ—Ç–≤–µ—Ç–∏–ª –ø—É—Å—Ç–æ. –ü—Ä–æ–±—É–µ–º fallback...")
        keys = fallback(seeds)
        WORDSTAT_RAW = "\n".join(f"{k}\t0" for k in keys)
    else:
        df = df.drop_duplicates("keyword").head(TOP_N_KEYWORDS)
        WORDSTAT_RAW = "\n".join(f"{row.keyword}\t{int(row.shows)}" for row in df.itertuples(index=False))
        print("‚úÖ –ü–æ–¥–æ–±—Ä–∞–Ω—ã –∫–ª—é—á–∏ Wordstat:")
        print(df.head(15).to_string(index=False))

except Exception as e:
    print("‚ùå –û—à–∏–±–∫–∞ Wordstat:", e)
    # fallback, –Ω–æ —Ç–µ–ø–µ—Ä—å –≤—ã–≤–æ–¥–∏–º –ø—Ä–∏—á–∏–Ω—É
    keys = fallback(SEED_KEYWORDS)
    WORDSTAT_RAW = "\n".join(f"{k}\t0" for k in keys)
    print("‚Üí –ò—Å–ø–æ–ª—å–∑—É–µ–º fallback –∫–ª—é—á–∏ —Å –ø–æ–∫–∞–∑–∞–º–∏ 0.")
```
